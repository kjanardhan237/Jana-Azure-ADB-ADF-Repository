```
rdd1=sc.textFile('dbfs:/jana/file1.csv')
```

```
rdd1.collect()
```

```
rdd2=rdd1.zipWithIndex().filter(lambda a:a[1]>4).map(lambda a:a[0].split(','))
```

```
rdd3=rdd2.collect()[0]
```

```
skipline=rdd2.first()
rdd2.filter(lambda a:a !=skipline).toDF(rdd3).show()
```

```
df=rdd2.filter(lambda a:a !=skipline).toDF(rdd3)
```
