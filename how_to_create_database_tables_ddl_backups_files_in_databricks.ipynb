```
def createdb(database):
        alltabels = spark.catalog.listTables('customers_db')
        f=open('/temp/jana/bkp_{}.sql'.format(database),'w')
        for t in alltabels:
            ddls = spark.sql('show create table {}.{};'.format(database,t.name))
            f.write(ddls.first()[0])
            f.write('\n')
        f.close()
```

```
from multiprocessing.pool import ThreadPool
```

```
alldb = [db.databaseName for db in spark.sql('show databases').collect()]
alldb
processes = ThreadPool(4)
processes.map(createdb,[database for database in alldb])
```
