```
from pyspark.sql.types import *
```

```
Schema=StructType([StructField('id',StringType(),True),
                 StructField('name',StringType(),True),
                 StructField('adult_id',StringType(),True),
                 StructField('education',StringType(),True),
                 StructField('room_number',StringType(),True),
                 StructField('marriage_status',StringType(),True),
                 StructField('admin_name',StringType(),True),
                 StructField('family_details',StringType(),True),
                 StructField('colours',StringType(),True),
                 StructField('gender',StringType(),True),
                 StructField('number_01',StringType(),True),
                 StructField('number_02',StringType(),True),
                 StructField('number_03',StringType(),True),
                 StructField('country',StringType(),True),
                 StructField('number_04',StringType(),True)
                     ])
```

```
df=spark.read.schema(Schema).csv('dbfs:/databricks-datasets/adult/',header=True)
```

```
display(df)
```

```
from pyspark.sql.window import Window
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
df = df.withColumn('adult_id',col('adult_id').cast(IntegerType()))\
     .withColumn('number_01',col('number_01').cast(IntegerType()))\
     .withColumn('number_02',col('number_02').cast(IntegerType()))\
     .withColumn('number_03',col('number_03').cast(IntegerType()))
```

```
df.printSchema()
```

```
display(df.drop('number_04'))
```

```
df.count()
```

```
df.dropDuplicates().count()
```

```
df_new = df.withColumn("rank",rank().over(Window.partitionBy("gender").orderBy(col("gender").desc().count())))\
.withColumn("dense_rank",dense_rank().over(Window.partitionBy("gender").orderBy(col("gender").desc().count())))\
.withColumn("row_number",row_number().over(Window.partitionBy("gender").orderBy(col("gender").desc().count())))\
.withColumn("ntile",ntile(2).over(Window.partitionBy("gender").orderBy(col("gender").desc().count())))\
.withColumn("percent_rank",percent_rank().over(Window.partitionBy("gender").orderBy(col("gender").desc().count())))
```
